```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the Titanic dataset
df = pd.read_csv("titanic.csv")
df.head()
```

```python
sns.set(style="whitegrid")
summary_stats = df.describe(include='all')
summary_stats
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
# Histograms for numeric features
plt.figure(figsize=(12, 6))
df.hist(bins=15, edgecolor='black', figsize=(12, 8))
plt.tight_layout()
plt.suptitle("Histograms of Numeric Features", fontsize=16, y=1.02)
plt.show()
```

```python
# Boxplots to detect outliers
plt.figure(figsize=(12, 6))
for i, column in enumerate(['Age', 'Fare'], 1):
    plt.subplot(1, 2, i)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass', 'SibSp']], hue='Survived', palette='Set2')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

- The distribution of passenger ages is right-skewed, with most passengers being in their 20s and 30s. There are fewer very young or very old passengers.
- Fare distribution is heavily right-skewed. Most passengers paid a lower fare, but there are a few outliers who paid significantly more, indicating the presence of first-class luxury tickets.
- No extreme outliers are observed in age, though a wide range is visible. This suggests a diverse age group among passengers.
